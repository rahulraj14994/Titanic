```python
#import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import itertools

import matplotlib.pylab as plt
%matplotlib inline 
```

```python
#load data
dfTitanic = pd.read_excel('Titanic.xlsx')
dfTitanic.head()
```

```python
# Find missing values in each column
missing_values = dfTitanic.isnull().sum()

# Print the missing values
print(missing_values)
```

```python
# Count the number of passengers in each passenger class
class_counts = dfTitanic['Passenger Class'].value_counts()

# Create a bar plot
class_counts.plot(kind='bar')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.title('Number of Passengers in Each Passenger Class')
plt.show()
```

```python
# Count the number of passengers in each class
passenger_count = dfTitanic['Passenger Class'].value_counts()

# Print the passenger count
print(passenger_count)
```

```python
# Count the number of passengers who survived
survived_counts = dfTitanic['Survived'].value_counts()

# Create a bar plot
survived_counts.plot(kind='bar')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Number of Survived Passengers')
plt.xticks([0, 1], ['No', 'Yes'])  # Customize the x-axis labels
plt.show()
```

```python
# Remove the 'Cabin' and 'Life Boat' columns
dfTitanic = dfTitanic.drop(['Cabin', 'Life Boat'], axis=1)

# Print the updated DataFrame
dfTitanic.head()
```

```python
# Calculate the average age
average_age = dfTitanic['Age'].mean()

# Fill missing age values with the average age
dfTitanic['Age'] = dfTitanic['Age'].fillna(average_age)

# Print the updated DataFrame
dfTitanic.head()
```

```python
# Remove rows with missing values
dfTitanic = dfTitanic.dropna()

# Print the updated DataFrame
dfTitanic.head()
```

```python
# Find missing values in each column
missing_values = dfTitanic.isnull().sum()

# Print the missing values
print(missing_values)
```

```python
# Count the number of passengers in each passenger class
class_counts = dfTitanic['Passenger Class'].value_counts()

# Create a bar plot
class_counts.plot(kind='bar')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.title('Number of Passengers in Each Passenger Class')
plt.show()
```

```python
# Count the number of passengers who survived
survived_counts = dfTitanic['Survived'].value_counts()

# Create a bar plot
survived_counts.plot(kind='bar')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Number of Survived Passengers')
plt.xticks([0, 1], ['No', 'Yes'])  # Customize the x-axis labels
plt.show()
```

```python
# Export dfTitanic to an Excel file
dfTitanic.to_excel('dfTitanic.xlsx', index=False)
```
